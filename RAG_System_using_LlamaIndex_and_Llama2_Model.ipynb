```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
from llama_index.core.prompts.prompts import SimpleInputPrompt
```

```python
# Goal: load pdfs and query those pdfs using hugging face libraries
#Loading PDFs from the data folder which I stored on drive

documents = SimpleDirectoryReader("/content/drive/MyDrive/AIBootCamp/AI Experiments/AZ-104-Labs").load_data()
```

```python
documents
```

```python
# Now you take this document and apply your model for text summarization or any query you want

system_prompt = """
You are a Q&A assistant. Your goal is to answer questions as
accurately as possible on the instructions and context provided.
"""

## we are going to use
## Default format which is supported by Llama2

query_wrapper_prompt = SimpleInputPrompt("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
!huggingface-cli login
```

```python
# How to call llama2 model from hugging face?
# for that we use torch

import torch

llm = HuggingFaceLLM(
    context_window = 4096,
    max_new_tokens= 256,
    generate_kwargs={"temperature": 0.0, "do_sample": False},
    system_prompt = system_prompt,
    query_wrapper_prompt= query_wrapper_prompt,
    tokenizer_name= "meta-llama/Llama-2-7b-chat-hf",
    model_name = "meta-llama/Llama-2-7b-chat-hf",
    device_map= "auto",
    #uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16, "load_in_8bit":True}
)
```

```python
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
from llama_index.core import ServiceContext
from langchain.embeddings import HuggingFaceEmbeddings

from llama_index.embeddings.langchain import LangchainEmbedding
```

```python
embed_model = LangchainEmbedding(HuggingFaceEmbeddings(model_name = "sentence-transformers/all-mpnet-base-v2"))
```

```python
from llama_index.core import Settings


Settings.llm = llm
Settings.embed_model = embed_model
```

```python
# a document summary index needs both an llm and embed model
# for the constructor
index = VectorStoreIndex.from_documents(
    documents, embed_model=embed_model, llm=llm
)
```

```python
# a vector store index only needs an embed model
#index = VectorStoreIndex.from_documents(
#    documents, embed_model=embed_model, transformations=transformations
#)
```

```python

# ... until you create a query engine
query_engine = index.as_query_engine(llm=llm)
```

```python
response1 = query_engine.query("What is Azure VM?")
```
